#Loop No. 1: failed. Save the very first line of each json file.

```python
import os, json
import pandas as pd


path_to_json = '/Users/shaomanlee/Documents/twitterscraper-master/data/tweets/'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]

len(json_files)
```

```python
# here I define my pandas Dataframe with the columns I want to get from the json
jsons_data = pd.DataFrame(columns=['fullname', 'html', 'Twitter_id', 'likes', 'replies', 'retweets', 'text', 'timestamp', 'url', 'user'])

# we need both the json and an index number so use enumerate()
for index, js in enumerate(json_files):
    with open(os.path.join(path_to_json, js)) as json_file:
        json_text = json.load(json_file)

        fullname = json_text[0]['fullname']
        html = json_text[0]['html']
        Twitter_id = json_text[0]['id']
        likes = json_text[0]['likes']
        replies = json_text[0]['replies']
        retweets = json_text[0]['retweets']
        text = json_text[0]['text']
        timestamp = json_text[0]['timestamp']
        url = json_text[0]['url']
        user = json_text[0]['user']
        # here I push a list of data into a pandas DataFrame at row given by 'index'
        jsons_data.loc[index] = [fullname, html, Twitter_id, likes, replies, retweets, text, timestamp, url, user]

# now that we have the pertinent json data in our DataFrame let's look at it
print(jsons_data.head())
```
